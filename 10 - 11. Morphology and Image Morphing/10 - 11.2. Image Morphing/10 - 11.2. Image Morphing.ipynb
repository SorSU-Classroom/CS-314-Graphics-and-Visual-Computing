```python
'''

style: Markdown

# Introduction to Image Morphing

``Image Morphing'' is the process of gradually converting one image into another. This process has
been used in the past two decades in the entertainment industry, but it is also very useful in image
classification algorithms.

There are two processes in image morphing: **warp** and **morph**. The **warp** process is the
transformation of the image to a new coordinate system, while the **morph** process is the gradual
transformation of the image into another image.
'''
```

```python
'''
# Warp

The **warp** process is the transformation of the image to a new coordinate system. This process 
requires ``fiducial points'' in the image, which are points that are used to define the transformation.
``Fiducial Points'' are the anchor points within the image that guides the warp. Commonly, these points
are located at important features such as corners and edges. In face algorithms, these points are located
at the eyes, nose, and mouth.
'''
```

```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Asset Root
asset_root = os.path.join(os.getcwd(), '../../assets')

# Image Path
image_path = os.path.join(asset_root, 'images', 'face_1.jpg')

# Read Image and convert to RGB
input_image = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)

# Display Both Image
plt.figure("Image Morphing", figsize=(10, 10))

plt.imshow(input_image)
plt.title("Original Image")
plt.axis('off')

plt.show()
```

```python
# Apply Warp

# Define Fiducial Points
fiducial_points = np.array([
    [0, 0], [0, 100], [100, 0], [100, 100]
])

# Define New Points
new_points = np.array([
    [0, 0], [0, 200], [200, 0], [200, 200]
])

# Compute Transformation Matrix
transformation_matrix = cv2.getPerspectiveTransform(fiducial_points.astype(np.float32), new_points.astype(np.float32))

# Apply Transformation
warped_image = cv2.warpPerspective(input_image, transformation_matrix, (200, 200))

# Display Both Image
plt.figure("Image Morphing", figsize=(10, 10))

plt.subplot(1, 2, 1)
plt.imshow(input_image)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(warped_image)
plt.title("Warped Image")
plt.axis('off')

plt.show()
```
